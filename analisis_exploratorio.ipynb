```python
import pandas as pd

#Creación de un DataFrame llamado "datos"
datos = pd.DataFrame({'Columna1':[10,20,30,40],
                      'Columna2':[1000,2000,3000,4000],
                      'Columna3':['perro ',' gato',' vaca ','caballo']})

datos #llamamos el DataFrame "datos"
```

```python
# Suma de todos los elementos de todas las columnas
datos.sum()
```

```python
# método sum() aplicado a la Columna1
datos['Columna1'].sum()
```

```python
# método sum() aplicado a la Columna1 y Columna2
datos[['Columna1','Columna2']].sum()
```

```python
# Valores máximos (.max()) y mínimos (.min()) de todas las columnas de un DataFrame
max = datos.max()
min = datos.min()
print(max)
print(min)
```

```python
# Funciones agregadas a un conjunto de datos
datos.agg(['sum','min','max'])
```

```python
# Metodos aplicados a columnas especificas
datos.agg({'Columna1':['sum','min','max'],'Columna2':['sum','min','max']})
```

```python
# Funciones aplicadas a cada elemento
# Definimos la funcion

def funcion (argumento):
    return argumento * 2

datos.agg(funcion)
```

```python
# Método .sort_values(), para ordenar una variable o columna
datos.sort_values('Columna1')
```

```python
# Método .sort_values(), para ordenar una variable o columna, de mayor # a menor
datos.sort_values('Columna1',ascending=False)
```
