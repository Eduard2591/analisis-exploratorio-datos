## Análisis descriptivo de los datos

```python
# Operaciones de la columna "Salario"
x= datos.Salario

type(x)
```

```python
# Media aritmética del DataFrame (excluyendo columnas no numéricas)
print(datos.mean(numeric_only=True))
```

```python
# Media aritmética
print(x.mean())
```

```python
# Mediana de todo el conjunto de datos
print(datos.median(numeric_only=True))
```

```python
# Valores máximo del DataFrame
datos.max()
```

```python
# Valores mínimos del DataFrame
datos.min()
```

```python
# Valor maximo y minimo de "Salario"
print("El salrio máximo es: ",x.max())
print("El salario mínimo es: ",x.min())
```

```python
# Moda del DataFrame
print(datos.mode())
```

```python
# Moda de una variable categorica
datos["Género"].mode()
```

```python
# Cuantiles de todas las Variables
print(datos.quantile([0, 0.25, 0.5, 0.75, 1], numeric_only=True))
```

```python
# Cuantiles
cuantiles = [0, 0.25, 0.5, 0.75, 1]
datos.Salario.quantile(cuantiles)
```

```python
# Varianza de todo el DataFrame
print(datos.var(numeric_only=True))
```

```python
# Varianza de una variable
datos.Salario.var()
```

```python
# varianza de la variable "Salario " por género
datos.groupby("Género")["Salario"].var()
```

```python
# Desviación estándar del DataFrame
print(datos.std(numeric_only=True))
```

```python
#  Desviación estándar de una variable
datos.Salario.std()
```

```python
# Coeficiente de variación
print(datos.Salario.std()/datos.Salario.mean())
```

## Medidas de asimetría (procedimientos)

```python
# Asimetría del DataFrame
print(datos.skew(numeric_only=True))
```

```python
# Asimetría de la variable salario
print(datos.Salario.skew())
```

```python
# Coeficiente de curtosis del DataFrame
print(datos.kurtosis(numeric_only=True))
```

```python
# Coeficiente de curtosis para la variable 'Salario'
print(datos.Salario.kurtosis())
```

### Para la variable Salario, se puede decir que la distribución de los datos es platicúrtica, dado que su coeficiente de curtosis es menor que cero. Esto indica una baja concentración de los datos respecto a la media aritmética.

```python
# Correlación para las variables numéricas del conjunto de datos
print(datos.corr(numeric_only=True))
```

### Se puede observar una alta correlación entre las cuatro variables (todas mayores a 0.7), dado que dependiendo del salario dependen los gastos, la capacidad de ahorro y el crédito aprobado.
